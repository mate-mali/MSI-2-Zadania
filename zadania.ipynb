# 1.9. Pewne schorzenie występuje u jednej na dziesięć tysięcy osób. Badanie wykrywające schorzenie daje wynik fałszywie pozytywny dla 2% przypadków, a fałszywie negatywny dla 1%. U losowo wybranej osoby wynik badania jest pozytywny (tzn. stwierdzono schorzenie). Oblicz prawdopodobieństwo, że ta osoba rzeczywiście jest chora.

```python

# prawdoposobienstwa napisane w treeasci zadania
prawd_chory = 1 / 10000 
prawd_zdrowy = 1 - prawd_chory  

# wskazniki false positives i false negatives, 
false_positive = 0.02  
false_negative = 0.01 

# Warunkowe prawdopodobieństwa pozytywnego wyniku - uzywalem formul z tych zrodel z drobna parafrazą https://pl.wikipedia.org/wiki/Twierdzenie_o_prawdopodobie%C5%84stwie_ca%C5%82kowitym, https://manuals.pqstat.pl/statpqpl:diagnpl:testpl
naprawde_chory  = 1 - false_negative 
naprawde_zdrowy  = false_positive     

# Prawo całkowitego prawdopodobieństwa: 
prawdopodobienstwo_poprawnej_choroby_diagnozy = naprawde_chory * prawd_chory + naprawde_zdrowy * prawd_zdrowy

# Twierdzenie Bayesa: 
prawdopodobienstwo_choroby_naprawde_chory = (naprawde_chory* prawd_chory) / prawdopodobienstwo_poprawnej_choroby_diagnozy

print(f"Prawdopodobienstwo ze jedna na dziesiec tysiecy osob bedzie zdiagnozowana jako chora i bedzie zdiagnozowana poprawnie = {prawdopodobienstwo_choroby_naprawde_chory :.5f}")
```

# 1.8. W pudełku są trzy monety: dwie zwykłe (orzeł i reszka) i jedna z orłem z obu stron. Wybieramy losowo jedną monetę i rzucamy jeden raz. Oblicz prawdopodobieństwo, że: (a) wypadnie orzeł, (b) wylosowano fałszywą monetę, jeżeli wiemy, że wypadł orzeł.

```python

# 3 monety: 2 zwykle (po 1 orle) + 1 falszywa (2 orly)

#(a) Prawdopodobienstwo wypadniecia orla 
prawd_wylosowania_monety = 1/3

prawd_orla_zwykla = 1/2   #na zwyklej monecie jest jeden orzel i jedna reszka 
prawd_orla_falszywa = 1   # falszywa moneta (orel z obu stron)

#z trzech monety mamy po rownej szansie ze wypadnie albo orzel albo reszka i takich monety mamy dwie, ale mamy tez 1.3 szansy ze trafimy na monete gdzie mamy 100% szansy na orla 
prawd_orla = prawd_wylosowania_monety * prawd_orla_zwykla + prawd_wylosowania_monety * prawd_orla_zwykla + prawd_wylosowania_monety * prawd_orla_falszywa 

print(f"a)Prawdopodobienstwo wylosowania orla : {prawd_orla:.4f}")

# (b) Prawdopodobienstwo ze wylosowano falszywa monete, wiedzac ze wypadl orel - twierdzenie Bayesa
prawd_falszywa_przy_orle = (prawd_orla_falszywa * prawd_wylosowania_monety) / prawd_orla

print(f"b) Prawdopodobienstwo ze wylosowano falszywa monete (wiedzac ze wypadl orel w pierwszym losowaniu): {prawd_falszywa_przy_orle:.4f}")
```

# 1.7. W każdym z trzech kontenerów jest 100 egzemplarzy pewnego produktu; w pierwszym jest 75 sprawnych i 25 uszkodzonych, w drugim 60 sprawnych i 40 uszkodzonych, a w trzecim 45 sprawnych i 55 uszkodzonych. Wybieramy losowo jeden  kontenerów, a następnie (również losowo) wybieramy z niego jeden egzemplarz. Oblicz prawdopodobieństwo, że wybrany egzemplarz jest sprawny.

```python
pojedynczy_kont_total = 100 #wszystkie maja 
kont1_uszkodzony = 25
kont2_uszkodzony = 40
kont3_uszkodzony = 55
prawd_wybrania_kontenera = 1/3 #(mamy trzy kontenery i kazdy moze byc wybrany z takim samym prawdopodobienstwem)

#jest 1/3 szansy ze wylosujemy dany kontener a potem ze wenatrz danego kontenera trafimy na sprawny egzemplarz, a ze w pospiechu zaczalem od zmiennych uszkodzonych to najpier policze te prawdopodobienstwa a potem sobie odwroce :D
prawd_uszkodzony = prawd_wybrania_kontenera * (kont1_uszkodzony/pojedynczy_kont_total) + prawd_wybrania_kontenera * (kont2_uszkodzony/pojedynczy_kont_total) + prawd_wybrania_kontenera * (kont3_uszkodzony/pojedynczy_kont_total)
prawd_sprawny = 1 - prawd_uszkodzony
print(f"Prawdopodobienstwo ze egzemplarz jest sprawny {prawd_sprawny:.2f}")
```

# 1.4. Oblicz prawdopodobieństwo śmierci w katastrofie lotniczej w ciągu dwudziestu lat dla osoby, która rocznie odbywa dwadzieścia podróży lotniczych. Załóż, że prawdopodobieństwo śmierci w katastrofie lotniczej wynosi pc = 1 /(4×10^6) .

```python
lotow_przez_20_lat = 20 * 20 #20 lotow w 20 lat
prawdopodobienstwo_smierci_jeden_lot = 1/(4*10**6)

przezycie_jednego_lotu = 1- prawdopodobienstwo_smierci_jeden_lot

prawdopodobienstwo_przezycia_wszystkich_lotow = (przezycie_jednego_lotu)**lotow_przez_20_lat

prawdopodobienstwo_smierci_w_ciagu_20_lat_lotow = 1 - prawdopodobienstwo_przezycia_wszystkich_lotow

print(f"Prawdopodobienstwo smierci w trakcie dwudziestu lat podrozowania samolotem: {prawdopodobienstwo_smierci_w_ciagu_20_lat_lotow:.9f}")
```

# 5.1. Oblicz 50! wykorzystując wzór Stirlinga

Wzor do odzwierciedlenia 
n!≈ 
(2πn)^(1/2) * ( e/n) ^ n

```python
from math import pi, e, sqrt #kilka funkcji z pakietu math, pi, liczba eulera, pierwiastek jesli potega do 1/2 nie zadziala 


def stirlinx(n):
    n_factorial = ((2 * pi * n) ** (1/2)) * ((n/e) ** (n))
    return n_factorial

print(f"50! uzywajac wzoru stirlinga wynosi {stirlinx(50)}")
```
