1.9. Pewne schorzenie występuje u jednej na dziesięć tysięcy osób. Badanie wykrywające schorzenie
daje wynik fałszywie pozytywny dla 2% przypadków, a fałszywie negatywny dla 1%. U
losowo wybranej osoby wynik badania jest pozytywny (tzn. stwierdzono schorzenie). Oblicz
prawdopodobieństwo, że ta osoba rzeczywiście jest chora.

```python

# prawdoposobienstwa napisane w treeasci zadania
prawd_chory = 1 / 10000 
prawd_zdrowy = 1 - prawd_chory  

# wskazniki false positives i false negatives, nie wiem jak to ladnie przetlumaczyc na polski wiec tak zostawiam 
false_positive = 0.02  
false_negative = 0.01 

# Warunkowe prawdopodobieństwa pozytywnego wyniku - uzywalem formul z tych zrodel z drobna parafrazą https://pl.wikipedia.org/wiki/Twierdzenie_o_prawdopodobie%C5%84stwie_ca%C5%82kowitym, https://manuals.pqstat.pl/statpqpl:diagnpl:testpl
naprawde_chory  = 1 - false_negative 
naprawde_zdrowy  = false_positive     

# Prawo całkowitego prawdopodobieństwa: 
p_plus = naprawde_chory * prawd_chory + naprawde_zdrowy * prawd_zdrowy

# Twierdzenie Bayesa: 
p_chory_given_plus = (naprawde_chory* prawd_chory) / p_plus

print(f"P(test+)              = {p_plus:.6f}")
print(f"P(chory | test+)      = {p_chory_given_plus:.6f}")
print(f"P(chory | test+) ≈ {p_chory_given_plus * 100:.4f}%")
```
