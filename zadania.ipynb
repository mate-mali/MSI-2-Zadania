wybieramy z niego jeden egzemplarz. Oblicz prawdopodobieństwo, że wybrany egzemplarz

5.1. Oblicz 50! wykorzystując wzór Stirlinga

Z tego co wyczytalem ten wzor wyglada tak 
n!≈ 
(2πn)^(1/2) * ( e/n) ^ n

```python
from math import pi, e, sqrt


def stirlinx(n):
    n_factorial = ((2 * pi * n) ** (1/2)) * ((n/e) ** (n))
    return n_factorial

print(f"50! uzywajac wzoru stirlinga wynosi {stirlinx(50)}")
```
